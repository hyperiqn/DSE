```python
import pandas as pd
import matplotlib as plt
import seaborn as sn
from sklearn.metrics import pairwise_distances 
from scipy.spatial.distance import cosine, correlation 
import numpy as np 
%matplotlib inline
```

```python
ratings = pd.read_csv("C:/Users/s_ani/Documents/da/ratings.csv")
ratings
```

### 2. Read the “ratings.csv” file and create a pivot table with index=‘userId’, columns=‘movieId’, values = “rating.

```python
pivot_table = ratings.pivot(index='userId', columns='movieId', values='rating')
pivot_table.fillna(0, inplace=True)
pivot_table
```

### 3. sklearn.metrics.pairwise_distances can be used to compute distance between all pairs of users. pairwise_distances() takes a metric parameter for what distance measure to use.

```python
distances = pairwise_distances(pivot_table.values, metric='cosine')
distances
```

### 6. Find the 5 most similar user for user with user Id 10.

```python
user_id = 10
similar_users = pd.Series(distances[user_id-1]).sort_values(ascending=True)[1:6].index + 1
print(f'most similar users are to user 10 are: {similar_users.to_list()}')
```

```python
movies = pd.read_csv("C:/Users/s_ani/Documents/da/movies.csv")
movies
```

### 7. Use the “movies” dataset to find out the names of movies, user 2 and user 338 have watched in common and how they have rated each one of them.

```python
merged_df=pd.merge(ratings, movies, on='movieId') 
```

```python
user2=merged_df[merged_df['userId'] == 2]
user2
```

```python
user338=merged_df[merged_df.userId==338]
user338
```

```python
common_movies = pd.merge(user2, user338, on='movieId')
common_movies
```

### 8. Use the movies dataset to find out the common movie names between user 2 and user 338 with least rating of 4.0

```python
common_movies_high_rating = common_movies[(common_movies['rating_x'] >= 4.0) & (common_movies['rating_y'] >= 4.0)]
common_movies_high_rating
```

### 9. Create a pivot table for representing the similarity among movies using correlation

```python
pivot_table = merged_df.pivot_table(index='userId', columns='title', values='rating')
correlation_table = pivot_table.corr(method='pearson')
correlation_table
```

### 10. Find the top 5 movies which are similar to the movie “Godfather”.

```python
godfather=correlation_table["Godfather, The (1972)"]
godfather.sort_values(ascending=False).head(5)
```
