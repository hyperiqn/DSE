```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/s_ani/Downloads/archive/players.csv")
df.head()
```

### Comparison of Market Value vs Position

```python
sns.boxplot(x='Position', y='Markey Value In Millions(£)', data=df)
plt.title('Market Value vs Position')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

From the graph, it is seen that second strikers have on average higher market value(34.9875 million). The range of market value for goalkeepers is the largest(16.2-63 million). Right Midfielders have the least market value(16.2-27 million)



### Comparison of Market Value vs Country

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Country', y='Markey Value In Millions(£)', data=df)
plt.title('Market Value vs Country')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

From the graph, its seen that Norway has the largest range of market value compared to other countries(16.2-135 million) though most players do not fall near the mean. Peru, Wales, Venezuela and Georgia all seem to have only 1 athlete each worth 16.2 million.

### Comparison of Market Value vs Age

```python
sns.boxplot(x='Age', y='Markey Value In Millions(£)', data=df)
plt.title('Market Value vs Age')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Here we see that players who are 18 have the highest average in terms of market value(46.95 million). However the players that are paid the highest lie in interval 21-30 years of age.

### Comparison of Market Value vs Club

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Club', y='Markey Value In Millions(£)', data=df)
plt.title('Market Value vs Club')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

Players in Inter Milan have on average the highest market value at 46.6875 million. However, the highest and second-highest worth players are in Paris Saint-Germain and Borussia Dortmund respectively.
