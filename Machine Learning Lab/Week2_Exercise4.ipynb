### 1. Multiple linear regression using builtin library on given dataset

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("multiple_linear_data.csv")
df
```

```python
X = df.drop('Y',axis= 1)
Y = df['Y']
Y = pd.DataFrame(Y)
```

```python
ax = plt.axes(projection ='3d')
ax.scatter3D(df['X1'], df['X2'], df['Y'], 'green')
ax.set_xlabel('X1', fontweight ='bold') 
ax.set_ylabel('X2', fontweight ='bold') 
ax.set_zlabel('Y', fontweight ='bold')
plt.show()
```

```python
reg = LinearRegression().fit(X, Y)
reg.intercept_ = reg.intercept_.round(2)
```

```python
print(f'{reg.intercept_[0]} + {reg.coef_[0][0]}*X1 + {reg.coef_[0][1]}*X2')
```

```python
X1 = np.linspace(1, 10, 200)
X2 = np.linspace(1, 10, 200)
Y_ = reg.intercept_[0] + reg.coef_[0][0]*X1 + reg.coef_[0][1]*X2
ax = plt.axes(projection ='3d')
ax.scatter3D(df['X1'], df['X2'], df['Y'], 'green')
ax.plot3D(X1, X2, Y_)
ax.set_xlabel('X1', fontweight ='bold') 
ax.set_ylabel('X2', fontweight ='bold') 
ax.set_zlabel('Y', fontweight ='bold')
plt.show()
```

```python
x1i = int(input('enter x1:'))
x2i = int(input('enter x2:'))
x = np.array([[x1i, x2i]])
y_pred = reg.predict(x)
print(f'predicted value:{y_pred}')
```

```python
ax = plt.axes(projection ='3d')
ax.scatter3D(df['X1'], df['X2'], df['Y'], 'green')
ax.scatter3D(x1i, x2i, y_pred)
ax.plot3D(X1, X2, Y_)
ax.set_xlabel('X1', fontweight ='bold') 
ax.set_ylabel('X2', fontweight ='bold') 
ax.set_zlabel('Y', fontweight ='bold')
plt.show()
```
