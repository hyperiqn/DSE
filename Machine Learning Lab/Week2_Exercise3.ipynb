### 1. Multiple linear regression on given dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("multiple_linear_data.csv")
df
```

```python
ax = plt.axes(projection ='3d')
ax.scatter3D(df['X1'], df['X2'], df['Y'], 'green')
ax.set_xlabel('X1', fontweight ='bold') 
ax.set_ylabel('X2', fontweight ='bold') 
ax.set_zlabel('Y', fontweight ='bold')
plt.show()
```

```python
df['X1X1'] = df['X1'] * df['X1']
df['X2X2'] = df['X2'] * df['X2']
df['X1X2'] = df['X1'] * df['X2']
df['X1Y'] = df['X1'] * df['Y']
df['X2Y'] = df['X2'] * df['Y']
df
```

```python
sum_y = df['Y'].sum()
sum_x1 = df['X1'].sum()
sum_x2 = df['X2'].sum()
sum_x1x1 = df['X1X1'].sum()
sum_x2x2 = df['X2X2'].sum()
sum_x1x2 = df['X1X2'].sum()
sum_x1y = df['X1Y'].sum()
sum_x2y = df['X2Y'].sum()
```

```python
sum_x1_2 = sum_x1x1 - sum_x1 * sum_x1/len(df)
sum_x2_2 = sum_x2x2 - sum_x2 * sum_x2/len(df)
sum_x1_y = sum_x1y - sum_x1 * sum_y/len(df)
sum_x2_y = sum_x2y - sum_x2 * sum_y/len(df)
sum_x1_x2 = sum_x1x2 - sum_x1 * sum_x2/len(df)
print(f'{sum_x1_2} {sum_x2_2} {sum_x1_y} {sum_x2_y} {sum_x1_x2}')
```

```python
b1 = (sum_x2_2 * sum_x1_y - sum_x1_x2 * sum_x2_y)/(sum_x1_2 * sum_x2_2 - sum_x1_x2**2)
b1 = b1.round(2)
b2 = (sum_x1_2 * sum_x2_y - sum_x1_x2 * sum_x1_y)/(sum_x1_2 * sum_x2_2 - sum_x1_x2**2)
b2 = b2.round(2)
a = df['Y'].mean() - b1 * df['X1'].mean() - b2 * df['X2'].mean()
a = a.round(3)
```

```python
print(f'Y = {a} + {b1}*X1 + {b2}*X2')
```

```python
X1 = np.linspace(0, 10, 10)
X2 = np.linspace(0, 10, 10)
Y = a + b1*X1 + b2*X2
ax = plt.axes(projection ='3d')
ax.scatter3D(df['X1'], df['X2'], df['Y'], 'green')
ax.plot3D(X1, X2, Y)
ax.set_xlabel('X1', fontweight ='bold') 
ax.set_ylabel('X2', fontweight ='bold') 
ax.set_zlabel('Y', fontweight ='bold')
plt.show()
```

```python
x1i = int(input('enter x1:'))
x2i = int(input('enter x2:'))
y_pred = a + b1*x1i + b2*x2i
print(f'predicted value:{y_pred}')
```

```python
ax = plt.axes(projection ='3d')
ax.scatter3D(df['X1'], df['X2'], df['Y'], 'green')
ax.scatter3D(x1i, x2i, y_pred)
ax.plot3D(X1, X2, Y)
ax.set_xlabel('X1', fontweight ='bold') 
ax.set_ylabel('X2', fontweight ='bold') 
ax.set_zlabel('Y', fontweight ='bold')
plt.show()
```
