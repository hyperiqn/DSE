## Week 5 - Exercise 3 - decision trees and random forest on heart dataset

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

### 1. decision tree

```python
df = pd.read_csv('RF_Dataset1_heart_v2.csv')
df.head()
```

```python
feature_cols = ['age','sex','cholestrol']
X = df[feature_cols]
y = df['heart disease']
y = pd.DataFrame(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf,feature_names=feature_cols,class_names='heart disease',filled=True)
```

```python
with open('dt1_heart.txt',"w") as f:
    f = tree.export_graphviz(clf, out_file=f, 
                                feature_names=feature_cols,  
                                class_names='heart disease',
                                filled=True)
```

### 2. random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X, y)
```

```python
tree.plot_tree(clf.estimators_[0],feature_names=feature_cols,class_names='heart disease',filled=True)
```

```python
with open('dt2_heart.txt',"w") as f:
    f = tree.export_graphviz(clf.estimators_[0], out_file=f, 
                                feature_names=feature_cols,  
                                class_names='heart disease',
                                filled=True)
```
