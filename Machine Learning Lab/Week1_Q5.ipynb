### f. Generate boxplots for each column to detect outliers

### g. Determine quartiles and remove outliers

```python
temp = df1.quantile([0.25, 0.5, 0.75]).T
temp["IQR"] = temp[0.75] - temp[0.25]
temp["min"] = temp[0.25] - temp["IQR"]*1.5
temp["max"] = temp[0.75] + temp["IQR"]*1.5

df1_f = pd.DataFrame(columns=df1.columns)

for i in range(df1.shape[0]):
    for j in range(df1.shape[1]):
        if (temp.iloc[j]["min"] <= df1.iloc[i, j] <= temp.iloc[j]["max"]) == False: break
    else: df1_f.loc[len(df1_f.index)] = df1.iloc[i]

df1_f
            
```

```python
temp = temp2.quantile([0.25, 0.5, 0.75]).T
temp["IQR"] = temp[0.75] - temp[0.25]
temp["min"] = temp[0.25] - temp["IQR"]*1.5
temp["max"] = temp[0.75] + temp["IQR"]*1.5

df2_f = pd.DataFrame(columns=temp2.columns)

for i in range(temp2.shape[0]):
    for j in range(temp2.shape[1]):
        if (temp.iloc[j]["min"] <= temp2.iloc[i, j] <= temp.iloc[j]["max"]) == False: break
    else: df2_f.loc[len(df2_f.index)] = temp2.iloc[i]

df2_f
```
