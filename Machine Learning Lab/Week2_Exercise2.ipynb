### 1. linear regression with inbuilt function on given dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df = pd.read_csv("dfregression.csv")
```

```python
df
```

```python
plt.scatter(df['x'], df['y'])
```

```python
reg = LinearRegression().fit(df[['x']],df['y'])
reg.intercept_ = reg.intercept_.round(2)
```

```python
print(f'{reg.intercept_} + {reg.coef_} * x')
```

```python
x = np.linspace(1, 10, 200)
y = reg.intercept_ + reg.coef_*x
plt.plot(x, y)
plt.scatter(df['x'], df['y'])
```

```python
xi = int(input("Enter an integer: "))
y_pred = reg.predict([[xi]])
print(f'the predicted value is:{y_pred}')
```

```python
plt.scatter(xi,y_pred)
plt.annotate('(%s, %s)' % (xi, y_pred), xy=(xi, y_pred), textcoords='offset points', xytext=(0,10), ha='center')
plt.plot(x, y)
plt.scatter(df['x'],df['y'])
for i, j in zip(df['x'], df['y']):
    plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
```
