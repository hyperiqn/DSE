```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
df = pd.DataFrame()
df = pd.read_csv("C://Users//s_ani//Documents//textbooks//4thSem//ML_LAB//Week3//datasets//winequality-red.csv", delimiter=";")
df.head()
```

### 1. Linear regression on winequality dataset

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
X = train['fixed acidity']
Y = train['pH']
```

```python
plt.scatter(X, Y)
```

```python
X = np.array(X).reshape((-1,1))
Y = np.array(Y).reshape((-1,1))
reg = LinearRegression().fit(X, Y)
```

```python
print(f'{reg.intercept_} + {reg.coef_} * x')
```

```python
x = np.linspace(0, 18, 100)
y = reg.intercept_ + reg.coef_*x
y = y.reshape(100,1)
```

```python
plt.scatter(X, Y)
plt.plot(x, y, c='red')
```

```python
xi = int(input("Enter an integer: "))
y_pred = reg.predict([[xi]])
print(f'the predicted value is:{y_pred}')
```

```python
plt.scatter(xi,y_pred,c='red')
plt.plot(x, y, c='red')
plt.scatter(X, Y, c='blue')
plt.scatter(test['fixed acidity'], test['pH'])
```

```python
predictions = reg.predict(np.array(test['fixed acidity']).reshape((-1,1)))
mse = mean_squared_error(test['pH'], predictions)
print(f'mse = {mse}')
```
